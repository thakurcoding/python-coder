```python
import smtplib
```

```python
import ssl
```

```python
from email.message import EmailMessage
```

```python
EMAIL= "Enter Your email"
APP_PASSWORD = "enter your app password"
RECEIVER = "Enter Receiver email"
```

```python
msg = EmailMessage()
```

```python
msg["from"] = EMAIL
msg["TO"] = RECEIVER
msg["subject"] = "hello guys this side Mahesh kumar talking to you......"
```

```python
msg.set_content("this email was shared by Mahesh to inform you that.....")
```

```python
context = ssl.create_default_context()
```

```python
with smtplib.SMTP_SSL("smtp.gmail.com",465,context=context)as server:
    server.login(EMAIL,APP_PASSWORD)
    server.send_message(msg)
```
