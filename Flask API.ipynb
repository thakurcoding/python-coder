```python
from flask import Flask
app = Flask(__name__)

@app.route("/")
def home():
    return """
    <html>
    <head>
        <title>My Simple Flask Page</title>
    </head>
    <body>
        <h1>Welcome to Flask!</h1>
        <p>This is a simple HTML page served using Flask.</p>
    </body>
    </html>
    """
if __name__=='__main__':
    app.run()

@app.route("/about")
def abouta():
    return """
    <html>
    <head>
        <title>My Simple Flask Page</title>
    </head>
    <body>
        <h1>Welcome to Flask!</h1>
        <p>This is a simple HTML page served using Flask.</p>
    </body>
    </html>
    """
if __name__=='__main__':
    app.run()

if __name__=='__main__':
    app.run(debug=True)
```
