```python
import pyttsx3 
import speech_recognition as sr


speaker = pyttsx3.init()
mic=sr.Recognizer()

speaker.say("Welcome To Mahesh")
speaker.runAndWait()
with sr.Microphone() as source :
    print("Start Speaking.....")
    voice=mic.listen(source)
    text=mic.recognize_google(voice)
    print(text)
```
